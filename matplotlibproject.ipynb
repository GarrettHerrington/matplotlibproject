```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sns.get_dataset_names()
```

```python
sns.get_dataset_names()
cc = sns.load_dataset('mpg')
cc.head()
```

```python
cc

#plt.scatter(cc['total'], cc['speeding'], cc['alcohol'])
plt.bar(cc['cylinders'], cc['mpg'])
x_label= "# Cylinders"
y_label= "MPG"
plt.ylabel(y_label)
plt.xlabel(x_label)
```

```python
plt.scatter(cc['displacement'], cc['horsepower'])
plt.xlabel("displacement")
plt.ylabel('horsepower')
```
